## INput data

```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
data = pd.read_csv('data/svm_data_200.csv')
data.head()
```

```python
fig = px.scatter_3d(data, x='x', y='y', z='z', color='class')
fig.show()
```

```python
a = [1, 0, 1.5]
b = [-1, 2, -2]
c = [0., 0.2, 1]
d = [1, -2, 2.5]
test_data = pd.DataFrame([a, b, c, d], columns=['x', 'y', 'z'])
```

## SOlution

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, f1_score
```

```python
X, y = data[['x', 'y', 'z']], data['class']
```

```python
clf = make_pipeline(StandardScaler(), 
                    SVC(C=1.1,
                        gamma='auto', 
                        kernel='rbf',
                        degree=6
                       ))
clf.fit(X, y)

train_predict = clf.predict(X)
acc = accuracy_score(y_pred=train_predict, y_true=y)
f1 = f1_score(y_pred=train_predict, y_true=y)

acc, f1
```

```python
predicted = clf.predict(test_data)
predicted
```

```python
data_to_plot = test_data.copy()
data_to_plot['class'] = predicted+2
data_to_plot = pd.concat([data, data_to_plot])
```

```python
fig = px.scatter_3d(data_to_plot, x='x', y='y', z='z', color='class')
fig.show()
```
